```python
import selenium
# import pyttsx3
from selenium import webdriver
import time
import os
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.action_chains import ActionChains
# import speech_recognition as sr
import pyautogui
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup
from collections import defaultdict
```

```python
options = Options()
options.add_argument("start-maximized")
options.add_argument("--disable-extensions")

options.add_experimental_option("prefs", { \
"profile.default_content_setting_values.media_stream_mic": 1,
"profile.default_content_setting_values.media_stream_camera": 1,
"profile.default_content_setting_values.geolocation": 1,
"profile.default_content_setting_values.notifications": 1
})
```

```python
def getprojectinfo(reflist):   
    
    
    driver = webdriver.Chrome(options=options,
                                  executable_path=ChromeDriverManager().install())

    information = defaultdict(list)

        # opening mahaonline webiee!

    driver.get("https://maharerait.mahaonline.gov.in/SearchList/Search")
    time.sleep(2)

        # clicking on the projectdetails

    driver.find_element_by_xpath('//*[@id="TheForm"]/div[1]/div[3]/div/div/div[2]/div[2]/a').click()
    time.sleep(2)

        ## click on registered projects

    driver.find_element_by_xpath('//*[@id="Agent"]').click()
    time.sleep(2)


        ## clicking on Advance button

    driver.find_element_by_xpath('//*[@id="btnAdvance"]').click()
    time.sleep(2)


        ## selecting element as Maharastra

    driver.find_element_by_xpath('//*[@id="State"]').send_keys('MAHARASTRA')
    time.sleep(2)


        ## Selecting the city as Mumbai 

    driver.find_element_by_xpath('//*[@id="District"]').send_keys('Mumbai City')
    time.sleep(2)


        # Click on the search

    driver.find_element_by_xpath('//*[@id="btnSearch"]').click()

    time.sleep(2)

    details = []
    
    window_before = driver.window_handles[0]

    for col in range(4):

            for i in range(0,len(reflist)):

                widow_before = driver.window_handles[0]

                driver.find_element_by_xpath(reflist[i]).click()

                time.sleep(1)


                current_tab=driver.window_handles[1]

                driver.switch_to.window(current_tab)


                try:

                    table = driver.find_element_by_xpath('//*[@id="DataGrid"]/div/div/table').text
                    table = table.split('\n')
                    tabcols = table[:-1]
                    tabdetails = table[-1]
                    tabdetails = tabdetails.split(' ')
                    details.append(tabdetails)




                except:

                     details.append(' ')


                driver.close()
                driver.switch_to_window(window_before)
                
                
    return details
  

# driver.switch_to_window(window_before)

# driver.switch_to_window(window_before)
# driver.find_element_by_xpath('//*[@id="btnNext"]').click()
# time.sleep(1)
```
